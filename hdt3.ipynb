## Analisis Expiatorio

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar los datos
datos = pd.read_csv('medicina.csv')
```

```python
# Codificación de características categóricas
datos_codificados = pd.get_dummies(datos)

# Para evaluar relaciones entre variables y calcular la correlación, podemos seleccionar solo las variables numéricas:
datos_numericos = datos_codificados.select_dtypes(include=[np.number])

# Exploración inicial de los datos
print("1. Visualizar las primeras filas:")
print(datos_numericos.head())
```

```python
print("\n2. Descripción básica de los datos:")
print(datos_numericos.describe())
```

```python
print("\n3. Distribución de características numéricas:")
datos.hist(figsize=(10, 8))
plt.show()
```

```python
print("\n4. Distribución de características categóricas:")
for column in datos.select_dtypes(include=['object']).columns:
    plt.figure(figsize=(8, 6))
    sns.countplot(x=column, data=datos)
    plt.show()
```

```python
print("\n5. Correlación entre variables numéricas:")
correlation_matrix = datos_numericos.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
print("\n6. Visualización de relaciones:")
sns.pairplot(datos_numericos)
plt.show()
```

```python
print("\n7. Manejo de valores faltantes:")
print(datos.isnull().sum())
```

Observaciones:

No hay valores faltantes en el conjunto de datos.
